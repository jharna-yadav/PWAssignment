```python
#Question 1 Demonstrate three different methods for creating identical 2D arrays in NumPy Provide the code for each
#           method and the final output after each method

lst = [[1,2,3], [1,2,3]]
#First method
arry1 = np.array(lst)
#Second method
array2 = np.zeros((2, 3), dtype=int)
array2[0, :] = [1, 2, 3]
array2[1, :] = [4, 5, 6]

#Third method
array3 = np.full((2, 3), 0, dtype=int)
array3[0, :] = [1, 2, 3]
array3[1, :] = [4, 5, 6]

print("First array : ", arry1)
print("Second array : ", array2)
print("Third array : ", array3)
```

          ***np.asarray***:-    It will reflect all the changes made to <br>original array.<br>

```python

# Question 14
from datetime import date, timedelta
from random import choice
def gen_random_dates(date1, date2):
  date_list =[]

  while(date1 <=date2):
    date1+= timedelta(days =1)
    date_list.append(date1.strftime('%Y-%m-%d'))

  return date_list
date1, date2 = date(2023, 6, 8), date(2023,7,27)
dates = gen_random_dates(date1, date2)
dates = np.array(dates)
dates
```

```python
#Question 17
# Generate two random arrays
np.random.seed(0)  # For reproducibility
x = np.random.randn(100)
y = np.random.randn(100)

# Determine the quadrant of each point
def get_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Quadrant I'
    elif x < 0 and y > 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

# Apply the function to each point
quadrants = [get_quadrant(xi, yi) for xi, yi in zip(x, y)]

# Create a DataFrame for Seaborn
df = pd.DataFrame({'x': x, 'y': y, 'quadrant': quadrants})

# Plot with Seaborn
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=df, x='x', y='y', hue='quadrant', palette='deep')

# Customize the plot
scatter_plot.set_title('Quadrant-wise Scatter Plot')
scatter_plot.set_xlabel('X-axis')
scatter_plot.set_ylabel('Y-axis')

# Add a legend
plt.legend(title='Quadrant')
plt.axhline(0, color='grey', linewidth=0.5)
plt.axvline(0, color='grey', linewidth=0.5)

# Show the plot
plt.show()
```

```python
#Question 18
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Ensure the plot will display in the notebook
output_notebook()

# Generate the x values
x = np.linspace(0, 2 * np.pi, 1000)

# Compute the sine of x
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave Function", x_axis_label='X-axis', y_axis_label='Y-axis', width=800, height=400)

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Add grid lines
p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'

# Show the plot
show(p)
```

```python
# Question 19
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import HoverTool
import numpy as np

categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 20, size=len(categories))

output_notebook()
p = figure(x_range=categories, height=350, title='Random Categorical Bar Chart',
           toolbar_location=None, tools='')

colors = ['#FF5733', '#33FF57', '#337AFF', '#FF33E9', '#FFD633']
p.vbar(x=categories, top=values, width=0.9, color=colors)

hover = HoverTool()
hover.tooltips = [('Value', '@top')]
p.add_tools(hover)

p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'

show(p)
```
